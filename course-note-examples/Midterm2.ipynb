```python
from symbulate import *
%matplotlib inline
```

```python
U1, U2 = RV(Uniform(0, 1) ** 2)
X = 2 + (5 - 2) * U1
Y = -X + (2 * X) * U2
```

```python
xy = (X & Y).sim(10000)
xy.plot()
xy.cov(), xy.corr()
```

```python
xy.plot('density')
```

```python
(X | (abs(Y - 3) < 0.01) ).sim(10000).plot()
```

```python
y = Y.sim(10000)
y.plot()
y.mean()
```
