```python
from symbulate import *
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
P = BoxModel({1:13, 0:39}, size = 5, replace = False)
X = RV(P, sum)
```

```python
x = X.sim(10000)
x.plot()
x.mean(), x.sd(), x.var()
```
