## Section 3 example

```python
X, U = RV(Normal(0, 1) * Uniform(-2, 2))
Y = X * U
```

```python
(Y | (abs(U - (-0.5)) < 0.01) ).sim(1000).plot()
```

```python
(X & Y).sim(1000).plot()
```

## Section 2 example

```python
X, Y = RV(Normal(100, 15) ** 2)

V = X + Y
W = (X & Y).apply(max)
```

```python
vw = (V & W).sim(1000)
vw.plot()
```

```python
vw.cov()
```
