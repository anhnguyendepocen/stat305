```python
from symbulate import *
%matplotlib inline
```

```python
P = Uniform(0, 1) ** 2
U1, U2 = RV(P)
```

```python
u = U1.sim(10000)
u.plot()
Uniform(0, 1).plot()
u.mean()
```

```python
W = 400 + 1200 * U1
w = W.sim(10000)
w.plot()
Uniform(400, 1600).plot()
w.mean()
```

```python
X = U1 + U2
x = X.sim(10000)
x.plot()
x.mean()
```

```python
Y = (U1 & U2).apply(max)
y = Y.sim(10000)
y.plot()
y.mean()
```

```python
Z = -log(1 - U1)
z = Z.sim(10000)
z.plot()
z.mean()
```

```python
Z = -log(U1)
z = Z.sim(10000)
z.plot()
z.mean()
```
