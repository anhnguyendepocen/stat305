```python
from symbulate import *
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
U1, U2 = RV(Uniform(0, 1) ** 2)
(U2 > 2 * U1).sim(10000).tabulate()
```

```python
(U1 & U2 | (U2 > 2 * U1) ).sim(1000).plot()
plt.xlim(0, 1)
```

```python
Z = 1 / (1 + U1)
Z.sim(10000).plot()
```

```python
X = (U1 & U2).apply(min)
Y = (U1 & U2).apply(max)
xy = (X & Y).sim(10000)
xy.plot('density')
```

```python
def fx(x):
    if x < 75:
        fx = x * 0.02 / 75
    else:
        fx = (100 - x) * 0.02 / 25
    return fx

U1, U2 = RV(Uniform(0, 1) ** 2)

X = (100 * U1 | ((100 * U1).apply(fx) > 0.02 * U2) )
```

```python
x = X.sim(10000)
x.plot()
```

```python
x.count_gt(65) / 10000
```
