```python
from symbulate import *
%matplotlib inline
from matplotlib import pyplot as plt
```

## Two spins of a Uniform(0, 1) spinner

```python
P = Uniform(0, 1) ** 2
U1, U2 = RV(P)
u1u2 = (U1 & U2).sim(100)
u1u2.plot()
u1u2
```

```python
(U1 & U2).sim(10000).plot()
```

```python
P = Uniform(0, 1) ** 2
X = RV(P, sum)
Y = RV(P, max)

xy = (X & Y).sim(100)
xy.plot()
xy
```

```python
(X & Y).sim(10000).plot('hist')
```

## Marginal distributions

```python
(X & Y).sim(10000).plot(['hist', 'marginal'])
```
