```python
from symbulate import *
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
P = BoxModel({1:13, 0:39}, size = 5, replace = True)
X = RV(P, sum)
```

```python
x = X.sim(10000)
x.plot()
x.mean(), x.sd(), x.var()
```

```python
P = BoxModel({1:13, 0:39}, size = 5, replace = True)
X = RV(P, sum)
X.sim(10000).plot()
Binomial(n = 5, p=0.25).plot()
```

```python
Hypergeometric(n = 5, N1 = 13, N0 = 39).pmf(5)
```

```python
Hypergeometric(n = 35, N1 = 17748, N0 = 21306 - 17748).plot()
Binomial(n = 35, p = 17748 / 21306).plot()
plt.xlim(20, 35);
plt.legend(["Hypergeometric","Binomial"]);
```

```python
X1, X2 = RV(Binomial(10, 0.3) * Binomial(5, 0.3))
(X1 + X2).sim(10000).plot()
Binomial(10 + 5, 0.3).plot()
```

```python
X1, X2 = RV(Binomial(10, 0.3) * Binomial(5, 0.3))
Y = X1 + X2
(X1 | (Y == 6) ).sim(10000).plot()
Hypergeometric(n = 10, N1 = 6, N0 = 10 + 5 - 6).plot()
```
