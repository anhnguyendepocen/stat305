# SAT Math scores from two spins of a Normal(500, 100) spinner - not realistic

```python
Q = Normal(500, 100) ** 2
X, Y = RV(Q)

xy = (X & Y).sim(100)
xy.plot()
xy
```

```python
(X & Y).sim(10000).plot()
```

```python
(X & Y).sim(10000).plot('hist')
```

# Bivariate Normal distributions - generate pairs of values

```python
P = BivariateNormal(mean1=527, mean2=533, sd1=107, sd2=100, corr=0.77)
P.sim(5)
```

```python
X, Y = RV(P)
(X & Y).sim(100).plot()
```

```python
(X & Y).sim(10000).plot()
```

```python
(X & Y).sim(10000).plot('hist')
```

```python
(X & Y).sim(10000).plot('density')
```

```python
X.sim(10000).plot()
```

```python
Y.sim(10000).plot()
```

# Changing the correlation

```python
P = BivariateNormal(mean1=527, mean2=533, sd1=107, sd2=100, corr=0.4)
X, Y = RV(P)

(X & Y).sim(10000).plot()
```

```python
P = BivariateNormal(mean1=527, mean2=533, sd1=107, sd2=100, corr=0.0)
X, Y = RV(P)

(X & Y).sim(10000).plot()
```

```python
P = BivariateNormal(mean1=527, mean2=533, sd1=107, sd2=100, corr=-0.77)
X, Y = RV(P)

(X & Y).sim(10000).plot()
```
