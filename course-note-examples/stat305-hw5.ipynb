```python
from symbulate import *
%matplotlib inline
```

```python
P = Uniform(0, 1) ** 2
U1, U2 = RV(P)
X = 1 + 2 * U1
Y = X + X * U2
(X & Y).sim(10000).plot('density')
```

```python
(Y | (abs(X - 3) < 0.01) ).sim(1000).plot()
```

```python
(X | (abs(Y - 3) < 0.01) ).sim(10000).plot()
```

```python
Y.sim(10000).plot()
```

```python
X, Y = RV(BivariateNormal(mean1 = 10, sd1 = 15, mean2 = 5, sd2 = 3, corr = -0.6))
R = 0.6 * X + 0.3 * Y + 0.1 * 1
r = R.sim(10000)
r.plot()
r.mean(), r.var(), r.sd()
```
