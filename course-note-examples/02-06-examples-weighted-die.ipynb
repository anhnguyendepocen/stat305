```python
from symbulate import *
%matplotlib inline
from matplotlib import pyplot as plt
```

## Fair four-sided die

```python
P = BoxModel([1, 2, 3, 4])
X = RV(P)
X.sim(10000).plot()
```

## Weighted four-sided die: `probs` definition

```python
Q = BoxModel([1, 2, 3, 4], probs = [0.1, 0.2, 0.3, 0.4])
X = RV(Q)
X.sim(10000).plot()
```

## Specify a random variable via its distribution

```python
X = RV(BoxModel([1, 2, 3, 4], probs = [0.1, 0.2, 0.3, 0.4]))
X.sim(10000).plot()
```

## Weighted four-sided die: Python "dictionary" definition

```python
P = BoxModel({1: 1, 2: 2, 3: 3, 4: 4})
X = RV(P)
X.sim(10000).plot()
```
