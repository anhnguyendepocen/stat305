## d) Covariance and variance of sums

Recall the SAT score examples from Lab 2. We assumed Math ($X$) and Reading
($Y$) scores had a *Bivariate Normal distribution* with mean 500 for each of Math and Reading, SD of 100 for each of Math and Reading, and a few different correlations. Consider total score $T = X + Y.$ **For each of the following cases, use formulas to find (1) $E(T)$ and (2) $SD(T)$, and compare to the simulated values.**  The code has been provided; you just need to run it.  In the plots of $T$ below, pay attention to the scale on the horizontal axis.

Case a) $Corr(X, Y) = 0$

```python
rho = 0 # correlation is often denoted with the small Greek letter rho
P = BivariateNormal(mean1 = 500, sd1 = 100, mean2 = 500, sd2 = 100, corr = rho)
X, Y = RV(P)
T = X + Y
(X & Y).sim(10000).plot('density')
```

```python
t = T.sim(10000)
t.plot()
t.mean(), t.sd()
```

Case b) $Corr(X, Y) = 0.8$

```python
rho = 0.8
P = BivariateNormal(mean1 = 500, sd1 = 100, mean2 = 500, sd2 = 100, corr = rho)
X, Y = RV(P)
T = X + Y
(X & Y).sim(10000).plot('density')
```

```python
t = T.sim(10000)
t.plot()
t.mean(), t.sd()
```

Case c) $Corr(X, Y) = 0.4$

```python
rho = 0.4
P = BivariateNormal(mean1 = 500, sd1 = 100, mean2 = 500, sd2 = 100, corr = rho)
X, Y = RV(P)
T = X + Y
(X & Y).sim(10000).plot('density')
```

```python
t = T.sim(10000)
t.plot()
t.mean(), t.sd()
```

Case d) $Corr(X, Y) = -0.8$

```python
rho = -0.8
P = BivariateNormal(mean1 = 500, sd1 = 100, mean2 = 500, sd2 = 100, corr = rho)
X, Y = RV(P)
T = X + Y
(X & Y).sim(10000).plot('density')
```
